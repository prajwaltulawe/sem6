```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('dataset_Facebook.csv',sep=";")
df
```

```python
#BoxPlot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Page total likes'])
plt.xlabel('Page total likes')
plt.title('Boxplot of Page total likes')
plt.show()
```

```python
#Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['Post Month'], bins=12)
plt.xlabel('Post Month')
plt.ylabel('Frequency')
plt.title('Histogram of Post Month')
plt.show()
```

```python
#Single Line Graph
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Page total likes'], marker='o')
plt.xlabel('Index')
plt.ylabel('Page total likes')
plt.title('Page total likes over time')
plt.show()
```

```python
df2=df[['Category']]
df2
```

```python
#MultiLine Graph
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Page total likes'], marker='o', label='Page total likes')
plt.plot(df2.index, df2['Category'], marker='o', label='Category')
plt.xlabel('Index')
plt.ylabel('Metrics')
plt.title('Multiple Metrics over time')
plt.legend()
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Page total likes', y='Lifetime Post Total Reach', hue='Type')
plt.xlabel('Page total likes')
plt.ylabel('Lifetime Post reach')
plt.title('Scatter Plot of Lifetime Total Post reach vs. Page total likes')
plt.show()
```

```python
#Bar Graph
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Post Month', y='Lifetime Engaged Users', hue='Category')
plt.xlabel('Post Month')
plt.ylabel('Lifetime Engaged Users')
plt.title('Bar Graph of Lifetime Engaged Users by Post Month and Category')
plt.show()
```

```python
#Box Plot with Notch
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Category', y='Post Month', notch=True)
plt.xlabel('Category')
plt.ylabel('Post Month')
plt.title('Box Plot with Notch of Post Month by Category')
plt.show()
```

```python
#pie Chart
plt.figure(figsize=(8, 8))
category_counts = df['Category'].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Categories')
plt.axis('equal')
plt.show()
```
